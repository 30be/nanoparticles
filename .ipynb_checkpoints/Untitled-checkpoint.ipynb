```python
from scipy import interpolate, integrate
import matplotlib.pyplot as plt
import colour
import pandas

def get_values(filename):
    return pandas.read_csv(filename, sep=r'\s+', header = None , skiprows = 3)

def get_color(a_name, r_name=None, water_name=None, frequencies=True, inverse=True, normalize=True):
    a = get_values(f"data/{a_name}")
    absorption = a[1]
    if r_name != None:
        absorption += get_values(f"data/{r_name}")[1]
    if water_name != None:
        absorption -= get_values(f"data/{water_name}")[1]
    if frequencies:
        a[0] = 300000 / a[0]
    wavelengths = a[0]
    if normalize:
        absorption /= max(absorption)
    if inverse:
        absorption = 1 - absorption
    return wavelengths, absorption
```

```python
cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
wavelengths = cmfs.wavelengths

red_perception = cmfs.values[:, 0] # Эффективность восприятия такой-то длинны волны колбочкой, ответственной за красный цвет 
green_perception = cmfs.values[:, 1]
blue_perception = cmfs.values[:, 2]
plt.plot(wavelengths, red_perception, color="red")
plt.plot(wavelengths, green_perception, color="green")
plt.plot(wavelengths, blue_perception, color="blue")
```

```python
def get_cie(measured_wavelengths, measured_spectra):
    # Спектр был от 400 до 750 нм, Натянем его на wavelengths, чтобы с ним совпадал(от 360 до 830) 
    spectra = interpolate.interp1d(measured_wavelengths, measured_spectra, kind = "linear", bounds_error = False, fill_value = 0)(wavelengths)
    
    # Умножим график восприятия на график полученного спектра, тем самым узнали сколько какого цвета было
    r = integrate.simpson(spectra * red_perception, wavelengths) # integrate(y(x), x)
    g = integrate.simpson(spectra * green_perception, wavelengths)
    b = integrate.simpson(spectra * blue_perception, wavelengths)
    return r / (r + g + b), g / (r + g + b)

colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)

for size in [25, 50, 75, 95]:
    cie = get_cie(*get_color(f"a{size}.txt", f"r{size}.txt"))
    plt.plot(*cie, 'o', label = size)

plt.xlabel('CIE x')
plt.ylabel('CIE y')
plt.legend()
plt.grid(True)
```

```python
for size in [25, 50, 75, 95]:
    plt.plot(*get_color(f"a{size}.txt", f"r{size}.txt", inverse=False), label=size)
plt.legend()
```

```python
# BE3
colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)
for i in range(1, 21):
    cie = get_cie(*get_color(f"ASC{i}.txt", f"RSC{i}.txt", frequencies=False))
    plt.plot(*cie, 'o', label = f"{i*10} нм")

plt.legend()
```

```python
measures = ["w", "Y", "Y1", "Y2", "G", "B",  "File_250514_172242", "File_250514_172903", "File_250514_173505",  "File_250514_175838"]
colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)
for f in measures:
    cie = get_cie(*get_color(f"14_05/{f}.txt", water_name=f"14_05/w.txt", frequencies=False))
    plt.plot(*cie, 'o', label = f)

plt.legend()
```
