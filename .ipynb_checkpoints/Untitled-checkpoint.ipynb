```python
import matplotlib.pyplot as plt

def get_values(filename="cross-sections.txt"):
    with open(filename) as f:
        lines = f.readlines()
        t = lambda x : [float(l.replace(',', '').split()[x]) for l in lines if not l.startswith("#")]
        return(t(0), t(1))

plt.plot(*get_values("cstg.txt"))
```

```python
import math

def g(x, mu, tau1, tau2):
  if x < mu:
    return math.exp(-(tau1**2) * (x - mu)**2 / 2)
  else:
    return math.exp(-(tau2**2) * (x - mu)**2 / 2)

def x_bar(lambda_val):
  term1 = 1.056 * g(lambda_val, 599.8, 0.0264, 0.0323)
  term2 = 0.362 * g(lambda_val, 442.0, 0.0624, 0.0374)
  term3 = -0.065 * g(lambda_val, 501.1, 0.0490, 0.0382)
  return term1 + term2 + term3

def y_bar(lambda_val):
  term1 = 0.821 * g(lambda_val, 568.8, 0.0213, 0.0247)
  term2 = 0.286 * g(lambda_val, 530.9, 0.0613, 0.0322)
  return term1 + term2

def z_bar(lambda_val):
  term1 = 1.217 * g(lambda_val, 437.0, 0.0845, 0.0278)
  term2 = 0.681 * g(lambda_val, 459.0, 0.0385, 0.0725)
  return term1 + term2

wavelengths = list(range(400, 700))
plt.plot(wavelengths, list(map(z_bar, wavelengths)))
plt.plot(wavelengths, list(map(x_bar, wavelengths)))
plt.plot(wavelengths, list(map(y_bar, wavelengths)))
```

```python
from scipy import interpolate
from scipy import integrate
import numpy as np
import colour
from colour.plotting import *
def intproduct(s, color, l):
    return integrate.simpson((s(l) * color), l)

cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
l = cmfs.wavelengths
a = cmfs.values[:, 0]
b = cmfs.values[:, 1]
c = cmfs.values[:, 2]

def find_coord (x,y):
    measurement_wavelengths, measurement_spectra = x,y
    spectra = interpolate.interp1d(measurement_wavelengths, measurement_spectra, kind="linear")
    c_x = intproduct(spectra, a, l)
    c_y = intproduct(spectra, b, l)
    c_z = intproduct(spectra, c, l)
    sum_xyz = c_x + c_y + c_z
    x = c_x / sum_xyz
    y = c_y / sum_xyz
    return x, y
plot_chromaticity_diagram_CIE1931(standalone=False)
def draw(x_coord, y_coord, label_str): # Renamed variables for clarity
    p = plt.plot(x_coord, y_coord,'x', label = label_str)
    if 0 <= x_coord <= 1 and 0 <= y_coord <= 1 and (1 - x_coord - y_coord) >= 0:
        plt.setp(p, 'markerfacecolor', (x_coord, y_coord, 1-x_coord-y_coord, 1))
    else:
         plt.setp(p, 'markerfacecolor', 'black')

def draw_color(filename):
    my_wavelengths, my_spectra = get_values(filename)
    my_cx, my_cy = find_coord(my_wavelengths, my_spectra)
    draw(my_cx, my_cy, "My Sample Spectrum")

    plt.title('CIE 1931 Chromaticity Diagram with Sample Point from' + filename)
    plt.xlabel('CIE x')
    plt.ylabel('CIE y')
    plt.legend()
    plt.grid(True)
```
