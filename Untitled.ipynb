```python
from scipy import interpolate
from scipy import integrate
import numpy as np
import colour
from colour.plotting import *
import matplotlib.pyplot as plt

def get_values(filename="cross-sections.txt"):
    with open(filename) as f:
        lines = f.readlines()
        t = lambda x : [float(l.replace(',', '').split()[x]) for l in lines if not l.startswith("#")]
        return(t(0), t(1))
plt.plot(*get_values("r75.txt"))
plt.show()
```

```python
def intproduct(s, color, l):
    return integrate.simpson((s(l) * color), l)

cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
l = cmfs.wavelengths
a = cmfs.values[:, 0]
b = cmfs.values[:, 1]
c = cmfs.values[:, 2]

def find_coord (x,y):
    measurement_wavelengths, measurement_spectra = x,y
    spectra = interpolate.interp1d(measurement_wavelengths, measurement_spectra, kind="linear", bounds_error=False, fill_value=0)
    c_x = intproduct(spectra, a, l)
    c_y = intproduct(spectra, b, l)
    c_z = intproduct(spectra, c, l)
    sum_xyz = c_x + c_y + c_z
    x = c_x / sum_xyz
    y = c_y / sum_xyz
    return x, y

def draw_color(filename):
    x_coord, y_coord = find_coord(*get_values(filename))
    plt.plot(x_coord, y_coord, 'o', label = filename)
    
plot_chromaticity_diagram_CIE1931(standalone=False)
for f in ["a25", "a50", "a75", "a95", "r25","r50", "r75", "r95"]:
    draw_color(f + ".txt")
plt.xlabel('CIE x')
plt.ylabel('CIE y')
plt.legend()
plt.grid(True)
plt.show()
```
