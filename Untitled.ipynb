```python
from scipy import interpolate, integrate
import matplotlib.pyplot as plt
import colour
import pandas

def get_values(filename):
    return pandas.read_csv(filename, sep='\t', header = None , skiprows = 3)

def get_normalized_absorption(size):
    absorption = get_values(f"data/a{size}.txt")
    scattering = get_values(f"data/r{size}.txt")
    actual_absorption = absorption[1] + scattering[1]
    return 300000 / absorption[0], actual_absorption / max(actual_absorption)
def get_color(size):
    wavelengths, absorption = get_normalized_absorption(size)
    return wavelengths, 1 - absorption
```

```python
cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
wavelengths = cmfs.wavelengths

red_perception = cmfs.values[:, 0] # Эффективность восприятия такой-то длинны волны колбочкой, ответственной за красный цвет 
green_perception = cmfs.values[:, 1]
blue_perception = cmfs.values[:, 2]
plt.plot(wavelengths, red_perception, color="red")
plt.plot(wavelengths, green_perception, color="green")
plt.plot(wavelengths, blue_perception, color="blue")
```

```python
def draw_color(size):
    measured_wavelengths, measured_spectra = get_color(size)
    
    # Спектр был от 400 до 750 нм, Натянем его на wavelengths, чтобы он с ним совпадал(от 360 до 830) 
    spectra = interpolate.interp1d(measured_wavelengths, measured_spectra, kind = "linear", bounds_error = False, fill_value = 0)(wavelengths)
    
    # Умножим график восприятия на график полученного спектра, тем самым узнали сколько какого цвета было
    r = integrate.simpson(spectra * red_perception, wavelengths) # integrate(y(x), x)
    g = integrate.simpson(spectra * green_perception, wavelengths)
    b = integrate.simpson(spectra * blue_perception, wavelengths)
    plt.plot(r / (r + g + b), g / (r + g + b), 'o', label = size)

colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)

for f in [25, 50, 75, 95]:
    draw_color(f)

plt.xlabel('CIE x')
plt.ylabel('CIE y')
plt.legend()
plt.grid(True)
```
