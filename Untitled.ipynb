```python
from scipy import interpolate, integrate
import matplotlib.pyplot as plt
import colour
import pandas

def get_values(filename):
    return pandas.read_csv(filename, sep=r'\s+', header = None , skiprows = 3)

def get_color(a_name, r_name=None, water_name=None, frequencies=True, inverse=True, normalize=True):
    a = get_values(f"data/{a_name}")
    absorption = a[1]
    if r_name != None:
        absorption += get_values(f"data/{r_name}")[1]
    if water_name != None:
        absorption -= get_values(f"data/{water_name}")[1]
    if frequencies:
        a[0] = 300000 / a[0]
    wavelengths = a[0]
    if normalize:
        absorption /= max(absorption)
    if inverse:
        absorption = 1 - absorption
    return wavelengths, absorption
```

```python
cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
wavelengths = cmfs.wavelengths

red_perception = cmfs.values[:, 0] #   -   ,     
green_perception = cmfs.values[:, 1]
blue_perception = cmfs.values[:, 2]
plt.plot(wavelengths, red_perception, color="red", label="()")
plt.plot(wavelengths, green_perception, color="green", label="()")
plt.plot(wavelengths, blue_perception, color="blue", label="()")
plt.xlabel(' , ')
plt.title("      ")
plt.legend()
```

```python
def get_cie(measured_wavelengths, measured_spectra):
    #    400  750 ,    wavelengths,    ( 360  830) 
    spectra = interpolate.interp1d(measured_wavelengths, measured_spectra, kind = "linear", bounds_error = False, fill_value = 0)(wavelengths)
    
    #       ,       
    r = integrate.simpson(spectra * red_perception, wavelengths) # integrate(y(x), x)
    g = integrate.simpson(spectra * green_perception, wavelengths)
    b = integrate.simpson(spectra * blue_perception, wavelengths)
    return r / (r + g + b), g / (r + g + b)

colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)

for size in [25, 50, 75, 95]:
    cie = get_cie(*get_color(f"a{size}.txt", f"r{size}.txt"))
    plt.plot(*cie, 'o', label = size)

plt.xlabel('CIE x')
plt.ylabel('CIE y')
plt.legend()
plt.grid(True)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()
for i in range(1, 21):
    color = get_color(f"ASC{i}.txt", f"RSC{i}.txt", frequencies=False, inverse=False)
    
    ax_index = int((i-1) / 5)
    axes[ax_index].plot(*color, label=f"{i*10} ")
    
    axes[ax_index].legend() 
    axes[ax_index].set_title(f" {(ax_index + 1)}")
    plt.xlabel(" , ")

fig.suptitle(" ")
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent overlap, leaving space for suptitle
```
