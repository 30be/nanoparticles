```python
from scipy import interpolate, integrate
import matplotlib.pyplot as plt
import colour
import pandas

def get_values(filename):
    return pandas.read_csv(filename, sep='\t', header = None , skiprows = 3)

def get_normalized_absorption(size):
    a = get_values(f"a{size}.txt")
    r = get_values(f"r{size}.txt")
    return 3*100000/a[0], (a[1]+r[1])/max(a[1]+r[1])
```

```python
cmfs = colour.colorimetry.MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
l = cmfs.wavelengths
a = cmfs.values[:, 0]
b = cmfs.values[:, 1]
c = cmfs.values[:, 2]
plt.plot(l, a)
plt.plot(l, b)
plt.plot(l, c)
plt.show()
```

```python
def intproduct(s, color, l):
    return integrate.simpson((s(l) * color), l)

def draw_color(size):
    measurement_wavelengths, measurement_spectra = get_normalized_absorption(size)
    spectra = interpolate.interp1d(measurement_wavelengths, measurement_spectra, kind="linear", bounds_error=False, fill_value=0)
    c_x = intproduct(spectra, a, l)
    c_y = intproduct(spectra, b, l)
    c_z = intproduct(spectra, c, l)
    sum_xyz = c_x + c_y + c_z
    plt.plot(c_x / sum_xyz, c_y / sum_xyz, 'o', label = size)

colour.plotting.plot_chromaticity_diagram_CIE1931(show=False)
for f in [25, 50, 75, 95]:
    draw_color(f)

plt.xlabel('CIE x')
plt.ylabel('CIE y')
plt.legend()
plt.grid(True)
plt.show()
```
